```python
def bar_chart(feature):
    left = df[df['Attrition']=='Yes'][feature].value_counts()
    still = df[df['Attrition']=='No'][feature].value_counts()
    dfr= pd.DataFrame([left,still])
    dfr.index= ['left','still']
    dfr.plot(kind='bar',stacked= True, figsize=(10,5))
```

```python
bar_chart('Gender')
```
